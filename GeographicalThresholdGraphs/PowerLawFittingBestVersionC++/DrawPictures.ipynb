```python
def draw_edges(filename, graphname):
    subplot(211)
    fin = open(filename, 'r')
    lines = fin.readlines()[1:]
    lines = [list(map(int, line.split(','))) for line in lines]
    vertes = [line[0] for line in lines]
    edges = [line[1] for line in lines]
    ex_edges = [line[2] for line in lines]
    title(graphname)
    plot(vertes, edges, label='practice', marker='o')
    plot(vertes, ex_edges, label='theory', marker='o')
    legend()
    subplot(212)
    import math
    errors = [ math.log(abs(edges[i] - ex_edges[i]) + 1) - math.log(vertes[i] * vertes[i-1]) for i in range(len(vertes))]
    plot(vertes, errors, label='errors', marker='o')
    legend()
```

```python
def draw_edges_one_plot(filename, graphname):
    import numpy as np
    import matplotlib.pyplot as plt
    
    
    fin = open(filename, 'r')
    lines = fin.readlines()[1:]
    lines = [list(map(int, line.split(','))) for line in lines]
    
    vertes = [line[0] for line in lines]
    edges = [line[1] for line in lines]
    ex_edges = [line[2] for line in lines]
    ex_sd = [line[3]**0.5 for line in lines]
    
    data = {}
    for i in range(len(vertes)):
        if vertes[i] in data:
            data[vertes[i]].append(edges[i])
        else:
            data[vertes[i]] = [edges[i]]
    import numpy as np
    ver = []
    mean = []
    var = []
    keys = data.keys()
    keys = sorted(keys)
    for i in keys:
        ver.append(i)
        mean.append(np.mean(data[i]))
        var.append(np.std(data[i], ddof=0))
    
    tn = ver
    ym = np.array(mean)
    ysd = np.array(var)
    

    fig, ax = plt.subplots()
    
    ax.plot(vertes, edges, 'go', label='practice', markersize=6)
    ax.plot(vertes, ex_edges, 'b--', label='theory', linewidth=2)
    
    np_ex_edges = np.array(ex_edges)
    np_ex_sd = np.array(ex_sd)
    
    ax.fill_between(tn, ym+ysd, ym-ysd, color = [0, 1, 0, 0.3])
    ax.fill_between(vertes, np_ex_edges+np_ex_sd, np_ex_edges-np_ex_sd, color = [0, 0, 1, 0.3])
    
    
    ax.set_xlabel('number vertexes')
    ax.set_ylabel('number edges')

    plt.title(graphname)

    # Hide the right and top spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    # Only show ticks on the left and bottom spines
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    
    plt.legend(loc='best', numpoints = 1)
```

```python
draw_edges_one_plot("alpha_1.txt", "alpha = 1")
```

```python
draw_edges_one_plot("alpha_2.txt", "alpha = 2")
```

```python
draw_edges_one_plot("alpha_3.txt", "alpha = 3")
```

```python
draw_edges("alpha_2.txt", "alpha = 2")
```

```python
draw_edges("alpha_3.txt", "alpha = 3")
```

```python
fin = open("triagles.txt", 'r')
lines = fin.readlines()[1:]
lines = [line.strip().split(',') for line in lines]
print lines
for i in range(len(lines)):
    if lines[i][1].strip() != '':
        lines[i] = list(map(int, lines[i]))
    else:
        lines[i] = list(map(int, lines[i][:1])) + [None] + list(map(int, lines[i][2:]))
```

```python
vertes = [line[0] for line in lines]
```

```python
real_triagles = [line[1] for line in lines if line[1] != None]
triagles = [line[2] for line in lines]
```

```python
title("number triagles")
plot(vertes[:len(real_triagles)-1], real_triagles[:len(real_triagles)-1], label='actual', marker='o')
plot(vertes[:len(real_triagles)-1], triagles[:len(real_triagles)-1], label='estimation', marker='o')
legend(loc=2)
```

```python
title("number triagles")
plot(vertes, triagles, label='estimation', marker='o')
plot(vertes, [2*x for x in vertes], label='y = 2*x')

legend(loc=2)
```

```python
title("log(number triagles)")
import math
plot(vertes, [math.log(x + 0.005) for x in triagles], label='estimation', marker='o')
plot(vertes, [math.log(x + 0.005) + math.log(2) for x in vertes], label='y = 2*x', marker='o')

legend(loc=4)
```

```python
"""
Demo of spines using custom bounds to limit the extent of the spine.
"""
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
y2 = y + 0.1 * np.random.normal(size=x.shape)

print len(x), len(y), len(y2)

fig, ax = plt.subplots()
ax.plot(x, y, 'k--')
ax.plot(x, y2, 'ro')

# set ticks and tick labels
ax.set_xlim((0, 2*np.pi))
ax.set_xticks([0, np.pi, 2*np.pi])
ax.set_xticklabels(['0', '$\pi$','2$\pi$'])
ax.set_ylim((-1.5, 1.5))
ax.set_yticks([-1, 0, 1])

# Only draw spine between the y-ticks
ax.spines['left'].set_bounds(-1, 1)
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

plt.show()
```
