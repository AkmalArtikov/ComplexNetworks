```python
fin_dates = open('cit-HepPh-dates.txt')
fin_edges = open('Cit-HepPh.txt')
```

```python
from datetime import date
dates = {}
data_to_vertexes = {}

for line in fin_dates:
    if len(line) > 0 and line[0] != '#':
        line = line.split()
        d = list(map(int, line[1].split('-')))
        d = date(d[0], d[1], d[2])
        dates[int(line[0])] = d
        if d in data_to_vertexes:
            data_to_vertexes[d] += 1
        else:
            data_to_vertexes[d] = 1
```

```python
edges = {}
data_to_edges = {}
for line in fin_edges:
    if len(line) > 0 and line[0] != '#':
        line = list(map(int, line.split()))
        edges[line[0]] = line[1]
        if line[0] in dates:
            data = dates[line[0]]
            if data in data_to_edges:
                data_to_edges[data] += 1
            else:
                data_to_edges[data] = 1
```

```python
dates_sorted = sort(data_to_vertexes.keys())
```
